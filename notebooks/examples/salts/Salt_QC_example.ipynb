# Salinity QC example

Example of Salinity QC from the Trolltransect2025 (not an example where the data quality was *excellent* -  more salinity samples should probably have been collected)

In this case, we load the following

- A log sample log sheet for the entire cruise: `Samplelog_TransectTokt_2025_pelagic_helium_and_salinity_samples.xlsx`
- Digitized salinometer readings: `TT25_salinometer_readings_digitized.xlsx`
- CTD bottle files: `00X.btl` w(we only prescribe the directory where all the `.btl` files are located)

### Locations of input files

```python
basedir = '../../../tests/test_data/salts/tt25/'
salts_sheet = f'{basedir}salts/TT25_salinometer_readings_digitized.xlsx'
log_sheet = f'{basedir}salts/Samplelog_TransectTokt_2025_pelagic_helium_and_salinity_samples.xlsx'
btl_dir = f'{basedir}btl/'
```

### Combine input files into a single xarray Dataset

```python
ds = salts.build_salts_qc_dataset(log_sheet, salts_sheet, btl_dir)
```

## Diagnosis plots

```python
salts.plot_salinity_diff_histogram(ds, 'PSAL1')
```

```python
salts.plot_salinity_diff_histogram(ds, 'PSAL2')
```
